# Cek Outlier

```python
df.describe()
```

### Cek Outlier

- `df.describe()`  
  Perintah ini menampilkan statistik deskriptif dari setiap kolom numerik, seperti:
  - `count`: jumlah data
  - `mean`: rata-rata
  - `std`: standar deviasi
  - `min`: nilai minimum
  - `25%`, `50%`, `75%`: kuartil 1 (Q1), median (Q2), kuartil 3 (Q3)
  - `max`: nilai maksimum

---

#### Analisis Outlier:

Salah satu cara mendeteksi **outlier** secara sederhana adalah dengan melihat **jarak antara nilai minimum/maksimum dengan Q1 dan Q3**. Jika nilai `min` atau `max` jauh dari kuartil, maka kemungkinan besar itu adalah outlier.

##### Contoh:

- **Kolom `charges`**  
  - Q1 = 4687.79  
  - Q3 = 16577.78  
  - Max = 63770.43  
  → Jarak `max` jauh di atas Q3, menunjukkan ada **outlier positif** (biaya medis sangat tinggi).

- **Kolom `bmi`**  
  - Q1 = 26.22  
  - Q3 = 34.77  
  - Max = 53.13  
  → Nilai `max` jauh dari Q3 → kemungkinan ada outlier BMI yang ekstrem.

- **Kolom `children`**  
  - Q3 = 2.0  
  - Max = 5  
  → Masih cukup wajar, tapi nilai `5` bisa dianggap **minor outlier** jika sangat jarang terjadi.

# Handling outlier

```python
df2 = df.copy()
```

- Membuat copy dari codingan df yang sudah pernah kita buat

```python
def capping(data) :
    upper = df2[x].quantile(0.90)
    lower = df2[x].quantile(0.10)
    df2[x] = np.where(df2[x] > upper, upper, 
                  np.where(df2[x] < lower, lower, df2[x]))
    print('upper ', upper, '&', 'lower', lower)
```

### Fungsi Capping (Pembatasan Outlier)

Fungsi `capping(data)` digunakan untuk **mengurangi pengaruh outlier ekstrem** pada suatu kolom numerik dalam DataFrame `df2` dengan cara membatasi nilainya pada rentang **persentil ke-10 hingga ke-90**.

---

#### Penjelasan Langkah-Langkahnya:

- **`upper = df2[x].quantile(0.90)`**  
  Menentukan batas atas, yaitu nilai pada persentil ke-90. Nilai di atas ini akan dianggap sebagai outlier dan "dipotong" (capped).

- **`lower = df2[x].quantile(0.10)`**  
  Menentukan batas bawah, yaitu nilai pada persentil ke-10. Nilai di bawah ini juga akan "dipotong".

- **`np.where(...)`**  
  Digunakan untuk mengganti nilai-nilai yang lebih tinggi dari `upper` menjadi `upper`, dan yang lebih rendah dari `lower` menjadi `lower`. Nilai dalam rentang normal tetap dipertahankan.

- **`print(...)`**  
  Menampilkan nilai batas atas dan batas bawah sebagai output agar pengguna tahu rentang yang digunakan untuk capping.




```python
kolom_numerik_sederhana = ['charges']
for x in kolom_numerik_sederhana :
    print(f'{capping(df2[x])} ini merupakan hasil dari perhitungan {x}')
    
```

#### Penjelasan Proses:

- List `kolom_numerik_sederhana` hanya berisi satu kolom, yaitu `'charges'`.
- Dalam `for` loop, setiap elemen (dalam hal ini `x = 'charges'`) akan diproses dengan memanggil fungsi `capping(df2[x])`.
- Nilai **upper** dan **lower** persentil dicetak dari dalam fungsi.
- Baris `print(f'{capping(df2[x])} ini merupakan hasil dari perhitungan {x}')` digunakan untuk memberikan keterangan hasil, namun mencetak `None` karena fungsi `capping()` tidak mengembalikan nilai (`return`).

Nilai MSE yang dihasilkan menjadi patokan dasar yang harus dikalahkan oleh model lain agar bisa dianggap lebih baik.
